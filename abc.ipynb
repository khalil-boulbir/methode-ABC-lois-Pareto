```python
import numpy as np
import pandas as pd

vmcum=[]
prcum=[]
order=[]
classe=[]
conteur=0
ind=[]
s=0
p=0
data=pd.read_csv("data.csv")
ind=data["Unnamed: 0"]
dat=data.drop("Unnamed: 0",axis=1)
matrix=dat.to_numpy()
dat=pd.DataFrame(matrix,index=ind,columns=["quantité","prix"])
print(dat)
dat["vm"]=dat["quantité"]*dat["prix"]
vm=dat["vm"].to_numpy()
x=np.sort(vm)
x=x[::-1]
dat=dat.sort_values(by=["vm"],ascending=False)
for i in x:
    s=s+i
    vmcum.append(s)
total=np.sum(vm)
for i in range(len(vmcum)):
    prcum.append((vmcum[i]/total)*100)
dat["vm"]=vm
dat["vm ordonne"]=x
dat["vm cummule"]=vmcum
dat["% cummule"]=prcum
for i in range(len(dat)):
    conteur=conteur+(100/len(dat))
    if(conteur<=20):
        classe.append("A")
    elif(conteur>20 and conteur<80):
        classe.append("B")
    else:
        classe.append("C")
    
dat["classement ABC"]=classe       
dat.to_csv("aaa.csv")
f=pd.read_csv("aaa.csv")
f["ordre d'importance"]=f["Unnamed: 0"]
f["product"]=ind
f=f.drop(["Unnamed: 0"],axis=1)
f.insert(6,"ordre des produit",f["ordre d'importance"])
f.insert(0,"product",f.pop("product"))
f=f.drop(["ordre d'importance"],axis=1)
f
```
